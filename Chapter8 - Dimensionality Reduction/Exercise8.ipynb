```python
# Imports

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import time # import time function to make time comparison
```

```python
# Loading the MNIST dataset

from sklearn.datasets import fetch_openml

mnist = fetch_openml('mnist_784', version = 1)
print(type(mnist))

X, y = mnist["data"], mnist["target"]
```

```python
# Splitting the data into train and test set

shuffle_index = np.random.permutation(60000)
X_train, y_train = X[shuffle_index], y[shuffle_index]
X_test, y_test = X[-shuffle_index], y[-shuffle_index]
```

```python
# Running a basic Random Forest

## Imports
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

## Instiating a RF object
classifier = RandomForestClassifier(random_state=42)

## Timing the fit
t0 = time.time()
classifier.fit(X_train, y_train)
t1 = time.time()
print("Training took {:.2f}s".format(t1 - t0))

## Checking the accuracy of the model with the test set
y_pred = classifier.predict(X_test)
print('Accuracy score', accuracy_score(y_test, y_pred))
```

```python
# Implementing PCA

## Imports
from sklearn.decomposition import PCA

## Instantiating PCA object with 95% variance explained
pca = PCA(n_components=0.95, random_state=42)

## Fitting and transforming the train data
X_train_reduced = pca.fit_transform(X_train)

## Instantiating basic RF object
clf = RandomForestClassifier(random_state=42)

## Timing the fit of RF after PCA transformation
t0 = time.time()
clf.fit(X_train_reduced, y_train)
t1 = time.time()
print("Training took {:.2f}s".format(t1 - t0))
```

```python
# Calculating the accuracy of the PCA->RF model with the test set

## Transforming the test data with PCA (not fitting!)
X_test_reduced = pca.transform(X_test)

## Predicting on the reduced dataset
y_pred = clf.predict(X_test_reduced)
print('Accuracy score', accuracy_score(y_test, y_pred))
```
