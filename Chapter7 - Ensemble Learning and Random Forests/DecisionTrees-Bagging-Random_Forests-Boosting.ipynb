# Decision Trees

```python
# Fitting a decision tree on iris data

## Imports
from sklearn.datasets import load_iris

## Loading the dataset
iris = load_iris()

## Only using the petal length and width as features
X = iris.data[:, 2:]

## Getting the target
y = iris.target
```

```python
# Training a DecisionTree Classifier

from sklearn.tree import DecisionTreeClassifier

clf = DecisionTreeClassifier(max_depth=2)
clf.fit(X, y)
```

```python
# Visualizing Decision Tree

from sklearn.tree import export_graphviz
from graphviz import Source
import os

## Settign up path 
PROJECT_ROOT_DIR = "/content"
CHAPTER_ID = "decision_trees"
IMAGES_PATH = os.path.join(PROJECT_ROOT_DIR, "images")
os.makedirs(IMAGES_PATH, exist_ok=True)

export_graphviz(
  clf,
  out_file=os.path.join(IMAGES_PATH, "iris_tree.dot"),
  feature_names=iris.feature_names[2:],
  class_names=iris.target_names,
  rounded=True,
  filled=True
)

Source.from_file(os.path.join(IMAGES_PATH, "iris_tree.dot"))
```

# Ensembling
* "wisdom of the crowd of classifiers"
* Where do you get these "crowd of classifiers"?
* using different algorithms to classify (trees, LR, SVM)
* use same model but resampling from the dataset with replacement

## Bagging
* Bootstrap Aggregating
* bootstrap: random sampling with replacement
* get a diverse set of classifiers by training on different random subsets of the data with replacement
* use mode for classification and average for regression

```python
# Training a Bagging Classifier

from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier

bag_clf = BaggingClassifier(
    DecisionTreeClassifier(),
    n_estimators=500,
    max_samples=100,
    bootstrap=True,
    n_jobs=-1
)

bag_clf.fit(X, y)
```

## Random Forests
* same hyperparameters as a DT
* same hyperparameters as Bagging
* chooses a random subset of features
* adds more diversity to each classifer/ regressor
* trades high bias for low variance

```python
# Training a Random Forest

from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(
    n_estimators=500,
)

rf.fit(X, y)
```
